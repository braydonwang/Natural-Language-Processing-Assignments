```python
import torch
import transformers
from transformers import BertTokenizer
```

```python
tokenizer = BertTokenizer.from_pretrained("bert-base-cased")
```

```python
tokenizer
```

```python
tokens = tokenizer.tokenize("A boy is playing a football")
```

```python
tokens
```

```python

encoded = tokenizer.encode("Hello world")
```

```python
encoded
```

```python
tokenizer.decode(encoded)
```

```python
encoded_plus = tokenizer.encode_plus("what is your name?", "John Doe!")
```

```python
encoded_plus
```

Sometimes token_type_ids are also refered as a segment_ids

```python
tokenizer.decode(encoded_plus['input_ids'])
```

## Set the max length

```python
encoded_plus = tokenizer.encode_plus("where are we?", "Planet Earth!", max_length=32, padding='max_length')
```
